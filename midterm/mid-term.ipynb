```python
import pandas as pd
jd = pd.read_csv('jumia file.csv')
jd.head()
```

```python
X= jd.drop(columns=['web-scraper-start-url','container_link-href','ratings-href'])
X.rename(columns={
        'web-scraper-order': 'N.o_order', 
        'container_link': 'Subject'
    }, inplace=True)
jd = X
jd.columns
```

```python
jd.isnull().sum()
```

```python
jd.head()
```

```python
jd['description'] = jd['description'].str.replace('\W', '', regex=True)
jd['ratings'] = jd['ratings'].str.replace('\W', '', regex=True)
jd['features'] = jd['features'].fillna('Not available', inplace=True)
```

```python
jd.head()
```

```python
jd.isnull().sum()
```

```python
jd[['Order_No','Method']] = jd['N.o_order'].str.split('-',expand=True)
```

```python
jd = jd.drop_duplicates(subset='Order_No', keep="first")
jd.head()
```

```python
jd['title'] = jd['title'].str.lower()
```

```python
jd['description'].dropna(inplace=True)
jd['description'] = jd['descrip'].astype(str)
jd.head()
```
